### import the libraries

```python
import requests
from bs4 import BeautifulSoup
```

### get request by the site

```python
r = requests.get("https://boxeoguide.com/search/name")
```

### give it to beautiful soup

```python
source = BeautifulSoup(r.content, "html.parser")
```

### get html tag 'td' ones

```python
tags = source.find_all("td")
```

```python
tags
```

### a function that removes html tags

```python
def remove_html_tags(text):
    """Remove html tags from a string"""
    import re
    clean = re.compile('<.*?>')
    return re.sub(clean, '', text).replace('\n', '').replace('\xa01\xa02', '').replace('*', '').lstrip()
```

### get img tags with class name 'profile-icon'

```python
source("img", attrs={'class':{'profile-icon'}})
```

### selecting male-female by src value

```python
sex_list = []
for value in source("img", attrs={'class':{'profile-icon'}}):
    if 'M-user_icon.png' in str(value):
        sex_list.append('Male')
    if 'F-user_icon.png' in str(value):
        sex_list.append('Female')
```

```python
sex_list
```

### column names

```python
column_names = ['Rating', 'Boxer', 'Country', 'Weight', 'Ceiling', 'Action', 'Promoter', 'Trainer', 'Sex']
```

### create dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=column_names)
```

```python
df
```

### fill in the 'Rating' column

```python
i = 0
for tag in tags:
    if i%9!=0:
        i+=1
        continue
    
    df.loc[i, 'Rating'] = int(remove_html_tags(str(tag)))
    i+=1
```
